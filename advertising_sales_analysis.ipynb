Q9

```python
#Dividing data into testing and training
from sklearn.model_selection import train_test_split
```

```python
#20% of data is used for testing and remaining 80% is used for training
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

Q11

```python
#Standardization
from sklearn.preprocessing import StandardScaler
```

```python
#fit is used to calculate mean and SD, transform is used to actually apply the changes
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
X_test
```

Q12

```python
#importing regression module from sklearn library
from sklearn.linear_model import LinearRegression
```

```python
#running the linear regression, n_jobs e
regression = LinearRegression(n_jobs = -1)
```

```python
regression.fit(X_train, Y_train)
LinearRegression(n_jobs=-1)
```

Q13

```python
#calculates coefficient and intercept with the linear regression model
print("Coefficient or Slope:", regression.coef_)
print("Intercept:", regression.intercept_)
```

Q14

```python
#using scatter plot to plot the line on the training data to see how well the line fits the data
plt.scatter(X_train,Y_train)
plt.plot(X_train,regression.predict(X_train))
```

Q15

```python
#predicting y on the test data
y_pred = regression.predict(X_test)
y_pred
```

Q16

```python
#import mse mae rmse modules from sklearn library
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
#computing errors to see the variability and test the accuracy of the prediction
mse = mean_squared_error(Y_test,y_pred)
mae = mean_absolute_error(Y_test,y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

Q17

```python
#scatter plot of actual vs predicted values to gauge model accuracy
plt.scatter(Y_test, y_pred)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red')
plt.xlabel("Actual Sales ($)")
plt.ylabel("Predicted Sales ($)")
plt.show()
```

Q18

```python
#plotting residuals to check if they are normally distributed
residuals = Y_test.values - y_pred
plt.hist(residuals, bins=20)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```

Q21

```python
#calculating r squared using sklearn and verifying it manually
from sklearn.metrics import r2_score
r2_sklearn = r2_score(Y_test, y_pred)
r2_sklearn
```

```python
#manual r squared calculation using SS_res and SS_tot
SS_res = np.sum((Y_test.values - y_pred) ** 2)
SS_tot = np.sum((Y_test.values - np.mean(Y_test)) ** 2)
r2_manual = 1 - (SS_res / SS_tot)
r2_manual
```
